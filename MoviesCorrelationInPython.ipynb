```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data
df = pd.read_csv(r'C:\Downloads\movies.csv')
```

```python
# Looking at the data
df.head()
```

```python
# Checking for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop all rows that have missing data
df = df.dropna()
```

```python
# Data types for the columns
df.dtypes
```

```python
# Change budget and gross columns data types from float64 to int64
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create a yearcorrect column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# View all the rows from now on
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates
df.drop_duplicates()
```

```python
df.head()
```

```python
# Budget will have high correlation to gross
# Company will have high correlation to gross
```

```python
# Scatterplot for budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Movie')
plt.show()
```

```python
df.head()
```

```python
# Plot for budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'black'}, line_kws={'color':'green'})
```

```python
# Looking at correlation
# Correlation function only works on numeric columns (integers or floats)
# Other correlation methods: pearson, kendall, spearman
df.corr(numeric_only=True, method='pearson')
```

```python
# High correlation between budget and gross
# 0.74
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Columns')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df_numerized.corr()
```

```python
# Changes the matrix above into a stacked version
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Sorts the previous code by lowest to highest correlation
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Returns only values that are > 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
